```python
print("=========TOKO KEL 13===========")
pembeli = input("Nama Pembeli : ")
print("Nama Pembeli :", pembeli)

def makanan():
    global totalmakanan
    global porsi
    global makan
    print("\n=============MENU MAKANAN=============")
    print("1. Nasi Kucing - Rp.4000")
    print("2. Nasi Uduk   - Rp.6000")
    print("3. Nasi Sayur  - Rp.8000")
    nomor = int(input("Masukan Pilihan 1/2/3 : "))
    porsi = int(input("Berapa Porsi : "))

    if nomor == 1:
        totalmakanan = porsi * 4000
        makan = "Nasi Kucing"
    elif nomor == 2:
        totalmakanan = porsi * 6000
        makan = "Nasi Uduk"
    elif nomor == 3:
        totalmakanan = porsi * 8000
        makan = "Nasi Sayur"
    else:
        print("Pilihan tidak ada! Silakan pilih kembali!")
        makanan()
    print(porsi, makan, "= Rp.", totalmakanan)

def minuman():
    global totalminuman
    global gelas
    global minum
    print("\n=============MENU MINUMAN=============")
    print("1. Pop Ice      - Rp.5000")
    print("2. Es Jeruk     - Rp.9000")
    print("3. Es Teh Manis - Rp.7000")
    nomor = int(input("Masukan Pilihan 1/2/3 : "))
    gelas = int(input("Berapa Gelas : "))

    if nomor == 1:
        totalminuman = gelas * 5000
        minum = "Pop Ice"
    elif nomor == 2:
        totalminuman = gelas * 9000
        minum = "Es Jeruk"
    elif nomor == 3:
        totalminuman = gelas * 7000
        minum = "Es Teh Manis"
    else:
        print("Pilihan tidak ada! Silakan pilih kembali!")
        minuman()
    print(gelas, minum, "= Rp.", totalminuman)

makanan()
minuman()

total_semua = totalmakanan + totalminuman

print("\nTotal yang harus dibayar : Rp.", total_semua)
uang = int(input("Uang Tunai Pembeli : Rp."))
kembalian = uang - total_semua
print("Kembalian : Rp.", kembalian)

print("\n============= S T R U K  B E L I =============")
print("Nama        :", pembeli)
print("Beli Makan  :", porsi, makan, "(Rp.", totalmakanan, ")")
print("Beli Minum  :", gelas, minum, "(Rp.", totalminuman, ")")
print("Tagihan     : Rp.", total_semua)
print("Dibayar     : Rp.", uang)
print("Kembalian   : Rp.", kembalian)
print("===============================================")
```
